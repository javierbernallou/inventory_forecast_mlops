```python
import pandas as pd 
import numpy as nd
```

```python
calendar = pd.read_csv(r"C:\Users\tela\Desktop\Javier_ING\portfolio\inventory_forecast_mlops\data\raw\calendar.csv")
sales = pd.read_csv(r"C:\Users\tela\Desktop\Javier_ING\portfolio\inventory_forecast_mlops\data\raw\sales_train_evaluation.csv")
sell_prices = pd.read_csv(r"C:\Users\tela\Desktop\Javier_ING\portfolio\inventory_forecast_mlops\data\raw\sell_prices.csv")
```

```python
print(f"Shapes of the datasets:")
print(f"Calendar: {calendar.shape}, Sales: {sales.shape}, Sell Prices: {sell_prices.shape}")
print(f"Calendar:\n{calendar.head()}")
print(f"Sales:\n{sales.head()}")
print(f"Sell Prices:\n{sell_prices.head()}")
```

Para este primer análisis, usaremos solo los datos de California, esto es debido a que en este notebook solo nos interesa conocer la compatibilidad de los datos, no unificarlos completamente ya que eso se hace en el archivo "preprocessing.py"

```python
sales = sales[sales["state_id"] == "CA"]
sell_prices = sell_prices[sell_prices["store_id"].str.startswith("CA")]
```

```python
id_columns = ["id", "item_id", "dept_id", "cat_id", "store_id", "state_id"]
for col in id_columns:
    sales[col] = sales[col].astype("category")
```

Procedemos a unir las tablas con melt:

```python
id_columns = ["id", "item_id", "dept_id", "cat_id", "store_id", "state_id"]
sales_melt = sales.melt(id_vars=id_columns, var_name="d", value_name="sales")
print(f"Filas originales: {sales.shape[0]} | Filas tras Melt: {sales_melt.shape[0]}")
```

```python
df = sales_melt.merge(calendar, on="d", how="left")
df = df.merge(sell_prices, on=["store_id", "item_id", "wm_yr_wk"], how="left")
```
